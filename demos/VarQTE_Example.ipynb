```python
from qflux.closed_systems.VarQTE import VarQITE, VarQRTE, ansatz_energy, Construct_Ansatz
```

### Define the Hamiltonian of interest

```python
H = SparsePauliOp.from_list([("IIZ", 1.0), ("IZI", 1.0), ("ZII", 0.65), ("IXX", 1.0), ("IYY", 1.0), ("XXI", 0.75), ("YYI", 0.75)])
```

### Create the initial state and run VarQITE

```python
qc = QuantumCircuit(2)
qc.rx(0.5, 0)
qc.rx(0.5, 1)

layers = 0
total_time = 10
timestep = 0.1
params = VarQITE(layers, H, total_time, timestep, init_circ=qc)

# Params now holds the parameter values for the ansatz at each timestep
for i in range(len(params)):
    print(f"Timestep {i} Energy: {ansatz_energy(qc, params[i], H)}")
```

### Create the initial state and run VarQRTE

```python
qc = QuantumCircuit(2)
qc.x(0) # Creates 100 state

layers = 1
total_time = 10
timestep = 0.1
params = VarQRTE(layers, H, total_time, timestep, init_circ=qc)
    
# Params now holds the parameter values for the ansatz at each timestep
for i in range(len(params)):
    print(f"Timestep {i} Energy: {ansatz_energy(qc, params[i], H)}")

# The ansatz circuit can be constructed as follows - this can be used to measure observables throughout time evolution
desired_timestep = 0
ansatz = Construct_Ansatz(qc, params[desired_timestep], H.num_qubits)
print(ansatz)
```
